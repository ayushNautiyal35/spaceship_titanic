```python
# for Destination
data['Destination'].dtype
```

```python
data['Destination'].value_counts()
```

```python
data['Destination']=data['Destination'].fillna('TRAPPIST-1e')
```

```python
data['VIP']=data['VIP'].fillna(False)
```

```python
data['VIP']=data['VIP'].map({False:0,True:1})
```

```python
#for RoomService
data['RoomService'].dtype
```

```python
data['RoomService'].describe()
```

```python
data['RoomService'].mode()
```

```python
data['RoomService']=data['RoomService'].fillna(0)
```

```python
#for FoodCourt
data['FoodCourt'].dtype
```

```python
data['FoodCourt'].describe()
```

```python
data['FoodCourt'].mode()
```

```python
data['FoodCourt']=data['FoodCourt'].fillna(0)
```

```python
#for ShoppingMall
data['ShoppingMall'].dtype
```

```python
data['ShoppingMall'].describe()
```

```python
data['ShoppingMall']=data['ShoppingMall'].fillna(0)
```

```python
#for Spa
data['Spa'].describe()
```

```python
data['Spa']=data['Spa'].fillna(0.0)
```

```python
#for VRDeck
data['VRDeck'].describe()
```

```python
#normalization of the data
data.info()
```

```python
for col in data.columns:
    if data[col].dtype=='float64':
        print(col)
```

```python
for col in data.columns:
    if data[col].max()>=10:
        print(col)
```

```python
#to normalize the float dtypes
float_dtype=['Age','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']

for i in float_dtype:
    maxi=max(data[i])
    mini=min(data[i])
    data[i]=(data[i]-mini)/(maxi-mini)
    
    
```

# 5. Feature Engineering

```python
import warnings
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
#by above some outliers are detected so removing them by iqr:
outliers=['Age','VIP','RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']
for i in outliers:
    q1=data[i].quantile(0.25)
    q3=data[i].quantile(0.75)
    iqr=q3-q1
    maximum=q3+1.5*iqr
    minimum=q1-1.5*iqr
    
    for j in range(0,len(data[i])):
        if data[i][j]>=maximum or data[i][j]<=minimum:
            data[i][j]=data[i].median()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, target, test_size=0.23, random_state=42)
```

# 6. Model Selection

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
```

```python
def check(model):
    y_pred=model.predict(x_test)
    print(f"Accuracy score= {accuracy_score(y_pred,y_test)}")
    print(f"Confusion Matrix = {confusion_matrix(y_pred,y_test)}")
    print(f"Precision score= {precision_score(y_pred,y_test)}")
    print(f"Recall score= {recall_score(y_pred,y_test)}")
    print(f"F1 score= {f1_score(y_pred,y_test)}")
```

```python
from sklearn.linear_model import LogisticRegression
check(LogisticRegression().fit(x_train,y_train))
```

```python
from sklearn.linear_model import PassiveAggressiveClassifier
check(PassiveAggressiveClassifier().fit(x_train,y_train))
```

```python
from sklearn.naive_bayes import GaussianNB
check(GaussianNB().fit(x_train,y_train))
```

```python
from sklearn.naive_bayes import BernoulliNB
check(BernoulliNB().fit(x_train,y_train))
```

```python
from sklearn.naive_bayes import MultinomialNB
check(MultinomialNB().fit(x_train,y_train))
```
